#Desafio Prático 1: Visualizando Dados Diferentes

###Criar um histograma da variável 'age' usando Matplotlib. (Base saude.csv)

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('saude.csv')
df.head()
```

```python
plt.hist(df['age'], bins=20, edgecolor='black')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.title('Histograma de Idade')
plt.show()
```

###Criar um gráfico de dispersão mostrando a relação entre 'height' e 'weight' usando Seaborn.

```python
import seaborn as sns

sns.scatterplot(x='height', y='weight', data=df)
plt.xlabel('Altura')
plt.ylabel('Peso')
plt.title('Relação entre Altura e Peso')
plt.show()
```

###Criar um gráfico interativo de barras mostrando a contagem de indivíduos por 'gender' usando Plotly.

```python
import plotly.express as px

contagem_genero = df['gender'].value_counts().reset_index()
contagem_genero.columns = ['Genero', 'contagem']
fig = px.bar(contagem_genero, x='Genero', y='contagem', title='Contagem de Indivíduos por Gênero')
fig.show()
```

#Desafio Prático 2: Séries Temporais

###Criar um gráfico de linhas das vendas ao longo do tempo usando Matplotlib. (Base: vendas.csv)

```python
df_vendas = pd.read_csv('vendas.csv')
df_vendas.head()
```

```python
df_vendas['date'] = pd.to_datetime(df_vendas['date'])
plt.plot(df_vendas['date'], df_vendas['sales'])
plt.xlabel('Data')
plt.ylabel('Vendas')
plt.title('Vendas ao Longo do Tempo')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

###Criar um gráfico de decomposição sazonal usando Seaborn

```python
!pip install statsmodels
import statsmodels.api as sm
```

```python
decomposition = sm.tsa.seasonal_decompose(df_vendas['sales'], model='additive', period=12)
fig = decomposition.plot()
fig.set_size_inches(10, 8)
plt.suptitle('Decomposição Sazonal das Vendas', y=1.02)
plt.show()
```

###Criar um gráfico interativo que permita a exploração dos dados de vendas ao longo do tempo usando Plotly.

```python
fig = px.line(df_vendas, x='date', y='sales', title='Vendas ao Longo do Tempo (Interativo)')
fig.show()
```
