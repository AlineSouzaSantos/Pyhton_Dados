Considerando a base de dados experimento_test_ab.csv responda as questões abaixo:
1. Qual dos cenários tem a maior taxa de conversão?
2. Calcule qual o tamanho da amostra necessária para o desenvolvimento de um teste A/B, seguindo os seguintes critérios:
a. O cenário A, da base, como o inicial, que funciona hoje.
b. Considere 95% de confiança de que o efeito na conversão não foi aleatório com um nível de significância de 5% (alpha = 0,05).
c. Também considere 80% de certeza conseguir capturar o efeito da nova abordagem.
d. O aumento para 10% de conversão.

```python
import pandas as pd
from statsmodels.stats.power import TTestIndPower

df = pd.read_csv('experimento_teste_ab.csv')
```

```python
df.head()
```

```python
#Não tem uma informação na base sobre o que seria o cenário, então deduzi que fosse a Versão_Página
taxa_conversao= df.groupby('Versão_Página')['Conversões'].mean()
cenario_maior_taxa = taxa_conversao.idxmax()
print(f"O cenário com a maior taxa de conversão é: {cenario_maior_taxa}")
print(f"Taxas de conversão por cenário:\n{taxa_conversao}")
```

```python
# 2. Calcule o tamanho da amostra necessária para o teste A/B

# a. Cenário A como inicial
taxa_conversao1 = df[df['Versão_Página'] == 'A']['Conversões'].mean()

# b. Nível de confiança de 95% e nível de significância de 5%
nivel_confianca = 0.95
alpha = 0.05

# c. Poder estatístico de 80%
poder = 0.80

# d. Aumento de 10% na conversão
aumento = 0.10
taxa_conversao2 = taxa_conversao1 * (1 + aumento)


analysis = TTestIndPower()
sample_size = analysis.solve_power(
    effect_size=aumento,
    power=poder,
    alpha=alpha,
    ratio=1
)
print(f"\nTamanho da amostra necessário por grupo: {int(round(sample_size))}")
```

```python
from scipy import stats

df_pacientes = pd.read_csv('pacientes.csv')
df_pacientes.head()
```

```python
df_pacientes = df_pacientes.drop(columns=['Unnamed: 0'])
df_pacientes.head()
```

```python
indices = [909, 751, 402, 400, 726, 39, 184, 269, 255, 769, 209, 715, 677, 381, 793, 697, 89, 280, 232, 756, 358, 36, 439, 768, 967, 699, 473, 222, 89, 639, 883, 558, 757, 84, 907, 895, 217, 224, 311, 348, 146, 505, 273, 957, 362]
amostra = df_pacientes.loc[indices, 'Idade']

# Teste t unilateral
t_stat, p_value = stats.ttest_1samp(amostra, 50, alternative='greater')

nivel_significancia = 0.05

print(f"Estatística t: {t_stat}")
print(f"Valor p: {p_value}")

if p_value < nivel_significancia:
    print("Rejeitamos a hipótese nula. A idade média das pessoas com problemas cardíacos é significativamente maior que 50 anos.")
else:
    print("Falhamos em rejeitar a hipótese nula. Não há evidências suficientes para afirmar que a idade média é maior que 50 anos.")
```

```python
print("\nQuestão 4:")
print("As amostras seriam independentes.  Dividir os pacientes em dois grupos (com e sem condições adicionais de saúde) cria dois grupos distintos e não relacionados entre si.")
```

```python
indices = [690, 894, 67, 201, 364, 19, 60, 319, 588, 643, 855, 623, 530, 174, 105, 693, 6, 462, 973, 607, 811, 346, 354,966, 943, 372]
amostra = df_pacientes.loc[indices]

grupo1 = amostra[amostra['Nome_Estado_Saude'] == 'Com condições de saúde adicionais']['Pressao_Arterial']
grupo2 = amostra[amostra['Nome_Estado_Saude'] == 'Saudável']['Pressao_Arterial']

# Teste t para amostras independentes
t_stat, p_value = stats.ttest_ind(grupo1, grupo2)

nivel_significancia = 0.06

print(f"\nQuestão 5:")
print(f"Estatística t: {t_stat}")
print(f"Valor p: {p_value}")

if p_value < nivel_significancia:
    print("Rejeitamos a hipótese nula. Há uma diferença significativa na pressão arterial média entre os grupos.")
else:
    print("Falhamos em rejeitar a hipótese nula. Não há evidências suficientes para afirmar que existe diferença significativa na pressão arterial média entre os dois grupos.")
```

```python
import statsmodels.formula.api as sm
import statsmodels.api as sm_base

# Definindo a fórmula para a ANOVA
formula = 'Pressao_Arterial ~ C(Etnia)'

# Executando o teste ANOVA
model = sm.ols(formula, data=df_pacientes).fit()
anova = sm_base.stats.anova_lm(model, typ=2)

print("\nQuestão 6:")
print(anova)

alpha = 0.05

if anova['PR(>F)'][0] < alpha:
    print("Rejeitamos a hipótese nula. Há uma diferença significativa na pressão arterial média entre pelo menos dois grupos étnicos.")
else:
    print("Falhamos em rejeitar a hipótese nula. Não há evidências suficientes para afirmar que existe uma diferença significativa na pressão arterial média entre os grupos étnicos.")
```

```python
tabela_contingencia = pd.crosstab(df_pacientes['Nome_Genero'], df_pacientes['Nome_Estado_Saude'])
chi2, p_value, dof, expected = stats.chi2_contingency(tabela_contingencia)

print("\nQuestão 7:")
print("Tabela de Contingência:")
print(tabela_contingencia)
print(f"Qui-quadrado: {chi2}")
print(f"Valor p: {p_value}")
print(f"Graus de liberdade: {dof}")

alpha = 0.05  # Nível de significância

if p_value < alpha:
    print("Rejeitamos a hipótese nula. Existe uma associação significativa entre sexo e condições de saúde adicionais.")
else:
    print("Falhamos em rejeitar a hipótese nula. Não há evidências suficientes para afirmar uma associação entre sexo e condições de saúde adicionais.")
```

```python
# Calcular a correlação de Pearson entre idade e pressão arterial
correlacao, p_valor = stats.pearsonr(df_pacientes['Idade'], df_pacientes['Pressao_Arterial'])

print("\nQuestão 8:")
print(f"Correlação de Pearson: {correlacao}")
print(f"Valor p: {p_valor}")

alpha = 0.05  # nível de significância

if p_valor < alpha:
    print("Rejeitamos a hipótese nula. Existe uma associação significativa entre idade e pressão arterial.")
else:
    print("Falhamos em rejeitar a hipótese nula. Não há evidências suficientes para afirmar uma associação entre idade e pressão arterial.")
```

```python
pressao_adicionais = df_pacientes[df_pacientes['Nome_Estado_Saude'] == 'Com condições de saúde adicionais']['Pressao_Arterial']
media = pressao_adicionais.mean()
desvio_padrao = pressao_adicionais.std()
n = len(pressao_adicionais)
intervalo = stats.t.interval(0.95, df=n-1, loc=media, scale=stats.sem(pressao_adicionais))

print("\nQuestão 9:")
print(f"Intervalo de confiança para a média da pressão arterial (pacientes com condições adicionais): {intervalo}")
```

```python
statistic, p_value = stats.shapiro(df_pacientes['Pressao_Arterial'])

print("\nQuestão 10:")
print(f"Estatística de Shapiro-Wilk: {statistic}")
print(f"Valor p: {p_value}")

alpha = 0.05

if p_value > alpha:
    print("Falhamos em rejeitar a hipótese nula. A distribuição da pressão arterial na população pode seguir uma distribuição normal.")
else:
    print("Rejeitamos a hipótese nula. A distribuição da pressão arterial na população não segue uma distribuição normal.")
```
