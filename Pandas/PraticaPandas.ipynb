```python
df_concurso = pd.read_csv('dados_concurso.csv')
df_concurso.head()
```

```python
#9 forma 1
df_concurso[df_concurso['Estado']=="PI"]['Número de Inscrição'].count()
```

```python
#9 forma 2
df_concurso[['Número de Inscrição','Estado']].groupby('Estado').count()
```

```python
#10 OBS dados não estão tratatos
filtro = (df_concurso['Estado']=="ES") & (df_concurso['Sexo']=='F')
qtd_fem_es = df_concurso[filtro]['Número de Inscrição'].count() 
qtd_fem_es
```

```python
#11 OBS dados não estão tratatos
df_concurso['Data de Nascimento'] = pd.to_datetime(df_concurso['Data de Nascimento'])
df_concurso[df_concurso['Data de Nascimento']<"1995-01-01"]['Número de Inscrição'].count()
```

```python
#11 Utilizando .loc
import numpy as np
df_antes95 = df_concurso.loc[df_concurso['Data de Nascimento'] < np.datetime64('1995')]
df_antes95.head()
```

```python
#12 - OBS dados não estão tratatos
from datetime import datetime
df_concurso['Data de Nascimento'] = pd.to_datetime(df_concurso['Data de Nascimento'])
ano_atual = datetime.today().year
df_concurso['Idade'] = ano_atual - df_concurso['Data de Nascimento'].dt.year
df_inscritos = df_concurso[['Número de Inscrição','Nome','Idade']]
df_inscritos
```

```python
#13 Quais foram os primeiros 10 inscritos?
df_10primeiras = df_concurso.set_index('Data de Inscrição').sort_index()[:10]
df_10primeiras
```

```python
#13 com .iloc
df_concurso.sort_values('Data de Inscrição').iloc[:10]
```
